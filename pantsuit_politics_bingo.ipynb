```python
# Parameters for plotting

xx = [0.5, 1.5, 2.5, 3.5, 4.5,
      0.5, 1.5, 2.5, 3.5, 4.5,
      0.5, 1.5,      3.5, 4.5,
      0.5, 1.5, 2.5, 3.5, 4.5,
      0.5, 1.5, 2.5, 3.5, 4.5]
yy = [4.5, 4.5, 4.5, 4.5, 4.5,
      3.5, 3.5, 3.5, 3.5, 3.5,
      2.5, 2.5,      2.5, 2.5,
      1.5, 1.5, 1.5, 1.5, 1.5,
      0.5, 0.5, 0.5, 0.5, 0.5]

blue = '#B6DBED'
red = '#FF99CB'
```

```python
def make_card(seed):
    np.random.seed(seed)
    select = np.random.choice(options, size=24, replace=False)

    plt.figure(figsize=(8,8), dpi=300)

    plt.xlim([0, 5])
    plt.ylim([0, 5])

    # Colored squares
    for xii in range(5):
        for yii in range(5):
            color_fill = red if ((xii + yii) % 2) else blue
            plt.fill(xii + np.array([0, 0, 1, 1]),
                     yii + np.array([0, 1, 1, 0]),
                     color_fill)
    
    # Grid lines
    for ii in range(0, 6):
        plt.axhline(ii, color='k')
        plt.axvline(ii, color='k')

    # Captions
    plt.annotate(free, xy=(2.5, 2.5),
                 horizontalalignment='center',
                 verticalalignment='center')

    for ii in range(24):
        plt.annotate(select[ii], xy=(xx[ii], yy[ii]),
                     horizontalalignment='center',
                     verticalalignment='center',
                     fontsize=8)

    # Remove weird ticks
    plt.xticks([])
    plt.yticks([])
    
    # Title
    plt.title("Pantsuit Politics Bingo", fontsize=20);
```

```python
interact(make_card, seed=(1, 1000, 1));
```
